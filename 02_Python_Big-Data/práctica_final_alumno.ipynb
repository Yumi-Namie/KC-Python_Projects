```python
from functools import reduce

'''
13. Dada una lista de Artículos con sus precio. Define las siguientes funciones:
Puedes definir más funciones si lo consideras necesario.
'''
articulos = {
    'Camisa': 20,
    'Pantalón': 30,
    'Calcetines': 5,
    'Zapatos': 50,
    'Gorra': 10,
    'Bufanda': 15,
    'Gafas': 25,
    'Reloj': 35,
    'Corbata': 40,
}

compra = ['Camisa', 'Pantalón', 'Pantalón', 'Gorra', 'Gafas', 'Corbata']


# A. Una función que calcule el precio total de la compra
#       Si quiero solo un valor (el total), uso el reduce.
#       La funcion me pide dos argumento: un acumulador y un valor. Doy una lista de compras y comprueba si hay en articulos y me añade a la suma del acumulador
total_compra = reduce(lambda total, item: total + articulos[item] if item in articulos else 0, compra, 0); print('Total compra:', total_compra)

# B. Una función que calcule el precio total de la compra con un descuento del 10%
#       Para calcular el valor total de la compra con descuento, necesito restar 10% del valor de la compra 
desc_10 = lambda total_compra:  total_compra - (total_compra*0.01) ; print('Total con desc de 10%:' , desc_10(total_compra))

# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€
#       Reutilizo la funcion anterior añadindo un condicional total de la compra mayor que 100
superior_100 = desc_10(total_compra) if total_compra > 100 else total_compra ; print('Compras superiores de 100: ', superior_100)

# D. Una función que calcule el precio total aplicando el IVA (21%)
#       Sumo el (valor de la compra) + ( total de la compra * 21%)
total_iva = lambda total_compra:  total_compra + (total_compra*0.21) ; print('Total con IVA sin descuento:', total_iva(total_compra))

# E.lista los artículos cuyo precio es superior a 20€
#       Si quiero una lista 'filtrada' de len != de la lista recebida, devo utilizar filter. 
#       Si los itens de compra está en articulos y su valor es mayor que 20, añado en la lista
superior_20 = list(filter(lambda item: True if item in articulos and articulos[item]>20 else False,compra)) ; print('Productos superiores a 20:', superior_20)
```
